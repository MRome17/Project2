```python
import pandas as pd

# List your files and corresponding seasons
file_years = [
    ('fangraphs-leaderboards.csv', 2021),
    ('fangraphs-leaderboards (1).csv', 2022),
    ('fangraphs-leaderboards (2).csv', 2023),
    ('fangraphs-leaderboards (3).csv', 2024),
    ('fangraphs-leaderboards (4).csv', 2025),
]
print(file_years)
```

```python
import pandas as pd

dfs = []

# Load each file, tag with the correct season
for filename, year in file_years:
    df = pd.read_csv(filename)
    df['Season'] = year
    dfs.append(df)

# Combine into one dataframe
combined_df = pd.concat(dfs, ignore_index=True)
combined_df.head()
```

```python
combined_df.to_csv("clutch_2021_2025_combined.csv", index=False)
```
